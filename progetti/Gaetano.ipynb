```python
import pandas as pd

# Carica il file Excel
df = pd.read_excel('/Users/andreacip/Downloads/tabella dati formula 1.xlsx')

# Assicurati che i nomi delle colonne nel DataFrame corrispondano a quelli qui sotto
colonne = ['qualifiche_sprint', 'sprint', 'gara', 'Capienza']

# Calcola la somma totale, la media, il minimo, il massimo e lo scarto quadratico medio
calcoli = {
    'Somma Totale': df[colonne].sum(),
    'Media': df[colonne].mean(),
    'Minimo': df[colonne].min(),
    'Massimo': df[colonne].max(),
    'Scarto Quadratico Medio': df[colonne].std()
}

# Crea un DataFrame dai calcoli
df_calcoli = pd.DataFrame(calcoli)

# Applica la formattazione per migliorare la leggibilit√† della tabella
df_stilizzato = df_calcoli.style.set_properties(**{
    'border-color': 'black',
    'border-style': 'solid',
    'border-width': '1px'
}).set_table_styles([{
    'selector': 'th',
    'props': [('font-size', '12pt')]
}])

# Visualizza la tabella stilizzata
df_stilizzato
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file Excel
df = pd.read_excel('/Users/andreacip/Downloads/tabella dati formula 1.xlsx')

# Lista delle colonne per cui generare i grafici, escludendo 'qualifiche'
colonne_grafici = ['qualifiche_sprint', 'sprint', 'gara', 'Capienza']

# Genera un grafico per ogni colonna
for colonna in colonne_grafici:
    plt.figure(figsize=(10, 5))  # Imposta la dimensione del grafico
    df[colonna].plot(kind='bar')  # Crea un grafico a barre
    plt.title(f'Grafico per {colonna}')  # Imposta il titolo del grafico
    plt.xlabel('Indice')  # Imposta l'etichetta dell'asse x
    plt.ylabel(colonna)  # Imposta l'etichetta dell'asse y
    plt.show()  # Mostra il grafico
```
