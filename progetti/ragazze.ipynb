```python
import pandas as pd

# Carica il file xlsx
df = pd.read_excel('/Users/andreacip/Downloads/indagine_studenti.xlsx')

# Calcoli statistici di base per ogni colonna
statistiche_descrittive = df.describe().round(0).astype(int)

# Stampa le statistiche descrittive in una tabella ben formattata
print("Statistiche Descrittive:")
display(statistiche_descrittive)

# Altri calcoli statistici
media = df.mean().round(0).astype(int)
mediana = df.median().round(0).astype(int)
moda = df.mode().iloc[0].round(0).astype(int)
varianza = df.var().round(0).astype(int)
deviazione_standard = df.std().round(0).astype(int)
coefficiente_di_variazione = (deviazione_standard / media).round(2)

# Crea un DataFrame per presentare i risultati in una tabella
risultati = pd.DataFrame({
    'Media': media,
    'Mediana': mediana,
    'Moda': moda,
    'Varianza': varianza,
    'Deviazione Standard': deviazione_standard,
    'Coefficiente di Variazione': coefficiente_di_variazione
})

# Stampa i risultati in una tabella ben formattata
print("Altri Calcoli Statistici:")
display(risultati)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assicurati che il percorso del file sia corretto
file_path = '/Users/andreacip/Downloads/indagine_studenti.xlsx'
df = pd.read_excel(file_path)

# Grafico 1: Studenti immatricolati per anno
plt.figure(figsize=(10, 6))
plt.plot(df['Anno'], df['Studenti_immatricolati'], marker='o', color='blue')
plt.title('Studenti immatricolati per Anno')
plt.xlabel('Anno')
plt.ylabel('Studenti Immatricolati')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.ticklabel_format(axis='y', style='plain')

# Grafico 2: Studenti in corso per anno
plt.figure(figsize=(10, 6))
plt.plot(df['Anno'], df['In_corso'], marker='x', color='green')
plt.title('Studenti in corso per Anno')
plt.xlabel('Anno')
plt.ylabel('In corso')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.ticklabel_format(axis='y', style='plain')

# Grafico 3: Totale positivi per anno
plt.figure(figsize=(10, 6))
plt.plot(df['Anno'], df['Laureati'], marker='s', color='orange')
plt.title('Laureati per Anno')
plt.xlabel('Anno')
plt.ylabel('Laureati')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.ticklabel_format(axis='y', style='plain')

# Grafico 4: Studenti fuori corso per anno
plt.figure(figsize=(10, 6))
plt.plot(df['Anno'], df['Fuori_corso'], marker='^', color='red')
plt.title('Studenti fuori corso per Anno')
plt.xlabel('Anno')
plt.ylabel('Fuori corso')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.ticklabel_format(axis='y', style='plain')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assicurati che il percorso del file sia corretto
file_path = '/Users/andreacip/Downloads/indagine_studenti.xlsx'
df = pd.read_excel(file_path)

# Imposta la dimensione del grafico
plt.figure(figsize=(10, 6))

# Aggiungi le serie di dati al grafico
plt.plot(df['Anno'], df['Studenti_immatricolati'], marker='o', color='blue', label='Studenti immatricolati')
plt.plot(df['Anno'], df['In_corso'], marker='x', color='green', label='Studenti in corso')
plt.plot(df['Anno'], df['Laureati'], marker='s', color='orange', label='Laureati')
plt.plot(df['Anno'], df['Fuori_corso'], marker='^', color='red', label='Studenti fuori corso')

# Aggiungi il titolo e le etichette degli assi
plt.title('Confronto tra Studenti Immatricolati, In Corso, Laureati e Fuori Corso per Anno')
plt.xlabel('Anno')
plt.ylabel('Numero di Studenti')

# Aggiungi la griglia, la legenda e formatta gli assi
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.ticklabel_format(axis='y', style='plain')

# Mostra il grafico
plt.show()
```
